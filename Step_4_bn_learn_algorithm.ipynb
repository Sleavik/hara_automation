```r
library(bnlearn)

options(jupyter.plot_mimetypes = 'image/png')

#reading csv file data, all columns' types are switched to "factor"
acc_data = read.csv("new_data.csv", colClasses=c(rep("factor",18)))
head(acc_data)
```

```r
#creating different types of graphs
list_from = read.csv("list_from.csv")
list_to = read.csv("list_to.csv")

bl = data.frame(from = c(list_from), to = c(list_to))
hc_bde_graph = hc(acc_data, score = "bde", blacklist = bl)
tabu_bic_graph = tabu(acc_data, score = "bic", blacklist = bl)
#gs_graph = gs(acc_data, test = "mi", alpha = , blacklist = bl)
#mmhc_graph = mmhc(acc_data, blacklist = bl)
#rs_graph = rsmax2(acc_data, restrict = "si.hiton.pc", test = "x2", maximize = "tabu", score = "bde", maximize = "tabu")
#pc_graph = pc.stable(acc_data, blacklist = bl)
#pc_graph2 = iamb.fdr(acc_data, undirected = FALSE, blacklist = bl)
```

```r
#scoring graphs
score(hc_bde_graph, data = acc_data, type = "bde")
score(tabu_bic_graph, data = acc_data, type = "bic")
#score(mmhc_graph, data = acc_data, type = "bde")

#rendering graph
#bde_graph = layoutGraph(bde_graph, attrs = list(graph = list(rankdir = "LR")))
#renderGraph (bde_graph)
```

```r
arc.strength(hc_bde_graph, data = acc_data, criterion = "mi")
```

```r
library(Rgraphviz) 
gr1 = graphviz.plot(hc_bde_graph, layout = "dot", shape = "rectangle", render = FALSE, highlight = list(arcs = rbind(incoming.arcs(hc_bde_graph, "Accident_Severity")), 
                                                                                                      col = "tomato", fill = "yellow", lwd = 2, lty = "dotted")) 
node.attrs = nodeRenderInfo(gr1)
node.attrs$fill["Accident_Severity"] = "tomato"
node.attrs$fill[ancestors(hc_bde_graph, "Accident_Severity")] = "yellow"
node.attrs$fill[parents(hc_bde_graph,"Accident_Severity")] = "orange"

nodeRenderInfo(gr1) = node.attrs 

renderGraph(gr1)
#graphviz.plot(pc_graph2, layout = "dot", shape = "rectangle")
#graphviz.plot(mmhc_graph, layout = "dot", shape = "rectangle")
#graphviz.plot(gs_graph, layout = "dot", shape = "rectangle")
#graphviz.plot(pc_graph, layout = "dot", shape = "rectangle")
```
