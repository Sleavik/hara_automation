```python
import random
import numpy as np
import matplotlib.pyplot as plt

# load the stats_17_UK dataset training data CSV file with accidents data into a list
accident_data_file = open("acc_data.csv", 'r')
accident_data_list = accident_data_file.readlines()
print(len(accident_data_list))
accident_data_file.close()

vehicle_data_file = open("veh_data.csv", 'r')
vehicle_data_list = vehicle_data_file.readlines()
print(len(vehicle_data_list))
vehicle_data_file.close()

# creating lists

dataset_dict = {}
acc_values = []
acc_values_temp = []
dataset_list = []
veh_values = []
veh_values_temp = []
exception_list =[]

# Accidents list content, separated by comma delimiter:
"""
0	Accident Index
1	Location Easting OSGR (Null if not known)
2	Location Northing OSGR (Null if not known)
3	Longitude (Null if not known)
4	Latitude (Null if not known)
5	Police Force
6	Accident Severity
7	Number of Vehicles
8	Number of Casualties
9	Date (DD/MM/YYYY)
10	Day of Week
11	Time (HH:MM)
12	Local Authority (District)
13	Local Authority (Highway Authority - ONS code)
14	1st Road Class
15	1st Road Number
16	Road Type
17	Speed limit
18	Junction Detail
19	Junction Control
20	2nd Road Class
21	2nd Road Number
22	Pedestrian Crossing-Human Control
23	Pedestrian Crossing-Physical Facilities
24	Light Conditions
25	Weather Conditions
26	Road Surface Conditions
27	Special Conditions at Site
28	Carriageway Hazards
29	Urban or Rural Area
30	Did Police Officer Attend Scene of Accident
31	Lower Super Output Area of Accident_Location (England & Wales only)
"""

# Vehicles list content, separated by comma delimiter:
"""
0	Accident Index
1	Vehicle Reference
2	Vehicle Type
3	Towing and Articulation
4	Vehicle Manoeuvre
5	Vehicle Location-Restricted Lane
6	Junction Location
7	Skidding and Overturning
8	Hit Object in Carriageway
9	Vehicle Leaving Carriageway
10	Hit Object off Carriageway
11	1st Point of Impact
12	Was Vehicle Left Hand Drive
13	Journey Purpose of Driver
14	Sex of Driver
15	Age of Driver
16	Age Band of Driver
17	Engine Capacity
18	Vehicle Propulsion Code
19	Age of Vehicle (manufacture)
20	Driver IMD Decile
21	Driver Home Area Type
"""
for i in range(1, len(vehicle_data_list)):
    # each veh_values_temp element is assigned accident_data_list whole line
    veh_values_temp += [vehicle_data_list[i].split(',')]
    # from each veh_values_temp element we extract only required sub elements and assign them to veh_values element
    veh_values = [c for k, c in enumerate(veh_values_temp[i - 1]) if k in {0, 2, 4, 7, 11}]   
    dataset_list.append(veh_values)

for i in range(1, len(accident_data_list)):
    # each acc_values_temp element is assigned accident_data_list whole line
    acc_values_temp += [accident_data_list[i].split(',')]
    # from each acc_values_temp element we extract only required sub elements and assign them to acc_values element
    acc_values = [c for k, c in enumerate(acc_values_temp[i - 1]) if k in {0, 6, 7, 8, 14, 16, 17, 18, 19, 20, 23, 24, 25, 26, 29}]
    dataset_dict[acc_values[0]] = acc_values[1:]

exceptioncount = 0
for j in range(len(dataset_list)):
    # adding data from acc_dict to dataset_list based on the keys in acc_dict.
    try:
        dataset_list[j].extend(dataset_dict[str(dataset_list[j][0])])
        dataset_list[j].pop(0)
    # If KeyError is raised then note added "NO ACCIDENT DATA"
    except KeyError:
        exceptioncount += 1
        exception_list.append(dataset_list[j])
        
#Counting number of erroneous data lines
print("There used to be " + str(exceptioncount) + " exceptions, that have been deleted successfully")
print(exception_list)

print(len(dataset_list))

data = np.asarray(dataset_list)
np.savetxt('veh+acc.csv', data, delimiter=',', fmt='%s')
```
